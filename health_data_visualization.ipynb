Data Visualization Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Line Chart

```python
df = pd.read_csv("Health_heart_experimental.csv")

# clean column names
df.columns = df.columns.str.strip()

hr_by_age = df.groupby("age")["HR"].mean()

plt.figure(figsize=(10,6))
plt.plot(hr_by_age.index, hr_by_age.values, marker='o')

plt.xlabel("Age (years)")
plt.ylabel("Average Heart Rate (bpm)")
plt.title("Average Heart Rate Across Different Ages")
plt.grid(True)
plt.show()

```

The line chart shows how average heart rate varies across age groups. A gradual change in heart rate can be observed as age increases, indicating age-related cardiovascular variation.

Bar Plot

```python
avg_weight = df.groupby("indication")["weightKg"].mean()

plt.figure(figsize=(6,5))
avg_weight.plot(kind="bar")

plt.xlabel("Health Indication")
plt.ylabel("Average Weight (kg)")
plt.title("Average Weight by Health Indication")
plt.show()
```

The bar plot compares the average body weight of healthy individuals with those at clinical risk.

Pie Chart

```python
indication_counts = df["indication"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    indication_counts,
    labels=["Healthy", "Clinical Risk"],
    autopct='%1.1f%%'
)
plt.title("Distribution of Health Indication")
plt.show()
```

The pie chart illustrates the proportion of healthy individuals versus those classified as having clinical risk.

Box Plot

```python
plt.figure(figsize=(6,5))
df.boxplot(column="DiaBP", by="indication")

plt.xlabel("Health Indication")
plt.ylabel("Diastolic Blood Pressure (mmHg)")
plt.title("Diastolic Blood Pressure by Health Indication")
plt.suptitle("")
plt.show()
```

The box plot highlights differences in the distribution of diastolic blood pressure between health groups and reveals possible outliers.

Histogram

```python
plt.figure(figsize=(7,5))
plt.hist(df["heightCm"], bins=10)

plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Height Distribution")
plt.show()
```

The histogram shows the frequency distribution of height values, indicating the most common height range in the dataset.

Scatter Plot

```python
plt.figure(figsize=(7,5))
plt.scatter(df["weightKg"], df["HR"])

plt.xlabel("Weight (kg)")
plt.ylabel("Heart Rate (bpm)")
plt.title("Relationship Between Weight and Heart Rate")
plt.show()
```

The scatter plot explores the relationship between body weight and heart rate, showing how heart rate varies with changes in body weight.
