## 1. Matrix Operations

```python
import numpy as np

A = np.array([[2, -1],
              [0,  3]])

B = np.array([[1,  2],
              [-3, 4]])

A + B, A @ B, 3*A - 2*B
```

## 2. Determinant and Inverse

```python
np.linalg.det(A), np.linalg.inv(A)
```

## 3. Eigenvalues and Eigenvectors

```python
C = np.array([[4, 1],
              [-2, 1]])

np.linalg.eig(C)
```

## 4. Linear Regression Using Normal Equation

```python
X = np.array([[4, 1.5, 100],
              [6, 2.0, 150]])

y = np.array([[20],
              [30]])

beta = np.linalg.pinv(X) @ y
beta
```

```python
x_new = np.array([[8, 3.0, 200]])
x_new @ beta
```
